```python
import matplotlib.pyplot as plt
import numpy as np
import cv2


#size: FFT        
size = 60

#thresh:          
thresh = 15

img = cv2.imread('image.jpg')
#img = cv2.blur(img,(10,10))
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")

print(img)
height , width = img.shape

print(height,width)


(x,y)=int(width/2.0),int(height/2.0)

print(x,y)

fft = np.fft.fft2(img)
fftShift = np.fft.fftshift(fft)
magnitude = 20 * np.log(np.abs(fftShift))

# display the original input image
(fig, ax) = plt.subplots(1, 2, )
ax[0].imshow(img, cmap="gray")
ax[0].set_title("Input")
ax[0].set_xticks([])
ax[0].set_yticks([])
# display the magnitude image
ax[1].imshow(magnitude, cmap="gray")
ax[1].set_title("Spectrum")
ax[1].set_xticks([])
ax[1].set_yticks([])
# show our plots
plt.show()
print(y-size,y+size,x-size,x+size)
fftShift[y - size:y + size, x - size:x + size] = 0
#fftShift[y - 60:y + 60, x - 60:x + 60] = 0
fftShift = np.fft.ifftshift(fftShift)
recon = np.fft.ifft2(fftShift)


magnitude = 20 * np.log(np.abs(recon))

plt.imshow(magnitude)
mean = np.mean(magnitude)

print(mean,mean<=thresh )
print(magnitude)
if(mean<= thresh):
    print("Blurry")
else:
    print("not Blurry")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


font = cv2.FONT_HERSHEY_SIMPLEX
cap = cv2.VideoCapture("testblur.mp4")
size = 60
thresh = 20

cap.get(cv2.CAP_PROP_FPS)

w = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('testbluroutput.mp4', fourcc, fps, (w, h))

while cap.isOpened():
    success,frame = cap.read()
    
    if success:
       
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        height , width = img.shape



        (x,y)=int(width/2.0),int(height/2.0)


        fft = np.fft.fft2(img)
        fftShift = np.fft.fftshift(fft)
        magnitude = 20 * np.log(np.abs(fftShift))

        
        fftShift[y - size:y + size, x - size:x + size] = 0
        fftShift = np.fft.ifftshift(fftShift)
        recon = np.fft.ifft2(fftShift)


        magnitude = 20 * np.log(np.abs(recon))

        mean = np.mean(magnitude)

        print(mean,mean<=thresh )
        
        if(mean<= thresh):
            print("Blurry")
        else:
            print("not Blurry")
            
        color = (0,0,255) if mean<=thresh else (0,255,0)
        text = " Blurry ({:.4f})" if mean<=thresh else "Not Blurry ({:.4f})"
        text = text.format(mean)
        cv2.putText(frame, text, (10, 25), font, 1, color, 2)
        out.write(frame)
        cv2.imshow('test',frame)
        
        if cv2.waitKey(25)&0xFF == 27: 
            break #   
            
    else :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
